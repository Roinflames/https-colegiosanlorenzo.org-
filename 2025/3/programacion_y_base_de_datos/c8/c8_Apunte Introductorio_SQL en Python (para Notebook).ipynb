1Ô∏è‚É£ ¬øQu√© es SQL?

SQL (Structured Query Language) es un lenguaje dise√±ado para consultar y manipular bases de datos relacionales.

Permite:

- Crear tablas ‚Üí CREATE TABLE

- Insertar datos ‚Üí INSERT

- Consultar informaci√≥n ‚Üí SELECT

- Actualizar informaci√≥n ‚Üí UPDATE

- Eliminar informaci√≥n ‚Üí DELETE

2Ô∏è‚É£ Usar SQL dentro de Python

Python incluye por defecto sqlite3, una base de datos liviana que funciona con archivos .db.

üîå Conexi√≥n a SQLite

```python
import sqlite3

# Crear o conectar a una base de datos
conn = sqlite3.connect("mi_base.db")

# Crear un cursor para ejecutar SQL
cursor = conn.cursor()
```

3Ô∏è‚É£ Crear una tabla con SQL

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS estudiantes (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nombre TEXT NOT NULL,
    edad INTEGER,
    carrera TEXT
)
""")

conn.commit()
```

4Ô∏è‚É£ Insertar datos

```python
cursor.execute("""
INSERT INTO estudiantes (nombre, edad, carrera)
VALUES ('Ana', 21, 'Ingenier√≠a Inform√°tica')
""")

cursor.execute("""
INSERT INTO estudiantes (nombre, edad, carrera)
VALUES ('Pedro', 23, 'Ingenier√≠a Civil')
""")

conn.commit()
```

5Ô∏è‚É£ Consultar datos (SELECT)

‚ñ∂Ô∏è Obtener todos los registros

```python
cursor.execute("SELECT * FROM estudiantes")
resultados = cursor.fetchall()

for fila in resultados:
    print(fila)
```

‚ñ∂Ô∏è Filtros con WHERE

```python
cursor.execute("""
SELECT nombre, edad
FROM estudiantes
WHERE edad > 21
""")

for fila in cursor.fetchall():
    print(fila)
```

‚ñ∂Ô∏è Ordenar resultados (ORDER BY)

```python
cursor.execute("""
SELECT nombre, edad
FROM estudiantes
ORDER BY edad DESC
""")

print(cursor.fetchall())
```

6Ô∏è‚É£ Usar SQL con pandas (m√°s c√≥modo)

```python
import pandas as pd

df = pd.read_sql_query("SELECT * FROM estudiantes", conn)
df
```

7Ô∏è‚É£ Actualizar y eliminar registros

üîÑ UPDATE

```python
cursor.execute("""
UPDATE estudiantes
SET edad = 24
WHERE nombre = 'Pedro'
""")

conn.commit()
```

‚ùå DELETE

```python
cursor.execute("""
DELETE FROM estudiantes
WHERE edad < 22
""")

conn.commit()
```

8Ô∏è‚É£ Cerrar la conexi√≥n

```python
cursor.execute("""
DELETE FROM estudiantes
WHERE edad < 22
""")

conn.commit()
```
